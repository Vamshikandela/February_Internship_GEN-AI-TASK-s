Problem 3: Find Maximum and Minimum Values


```python
numbers = [45, 22, 89, 10, 66]
max_num=numbers[0]
min_num=numbers[0]
print("List is:",numbers)
for number in numbers:
  if number>max_num:
    max_num=number
  if number<min_num:
    min_num=number
print("Max:",max_num)
print("Min:",min_num)
```


Problem 6: Remove Duplicate Phone Numbers


```python
phone_numbers = [9876543210, 9123456789, 9876543210]
unique=[]
for number in phone_numbers:
  if number not in unique:
    unique.append(number)
print("Unique phone_Numbers are: ",unique)
```

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
print("Unique phone_Numbers are:",set(phone_numbers))
```

7.Count Character Frequency

```python
# Take input from user
name=input("enter a name")
# Create an empty dictionary to store character count
freq={}
# Loop through each character in the input string
for ch in name:
# If character already exists in dictionary
  if ch in freq:
# If character already exists in dictionary, increase its count
    freq[ch]+=1
    #If not character already exists in dictionary initilaze count as 1
  else:
    freq[ch]=1

# Print the final frequency dictionary
print("\n Frequency of characters",freq)
```

Problem 9: Check if a Key Exists in a Dictionary


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000}
for name,salary in employees.items():
  if name=="Ravi":   # compares name with Ravi
    print(name,salary) # if matches return name ,salary
    break  # for loop will stops break will do
else:
  print("Employee is not exists")  # if not matches will give this


```
