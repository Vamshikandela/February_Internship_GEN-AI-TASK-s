**1.import Libraries**

```python
from langchain.prompts import PromptTemplate, ChatPromptTemplate
```

**2. Task 1: Replace Hardcoded Prompt**

```python
def create_basic_template():
    template = PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} in simple terms for beginners"
    )
    return template


def run_basic_template(topic):
    prompt = create_basic_template()
    return prompt.format(topic=topic)
```

```python
run_basic_template("Machine Learning")
```

**3. Task 2: Multi-Input Prompt System**





```python
def create_multi_input_template():
    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} tone"
    )
    return template


def run_multi_input(topic, audience, tone):
    prompt = create_multi_input_template()
    return prompt.format(topic=topic, audience=audience, tone=tone)
```

```python
print(run_multi_input("AI", "beginners", "friendly"))
print(run_multi_input("Python", "kids", "fun"))
print(run_multi_input("Deep Learning", "engineers", "technical"))
```

**4. Task 3: Prompt Variations Engine**

```python
def get_prompt_by_style(style):

    templates = {
        "teaching": PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} clearly step by step"
        ),
        "interview": PromptTemplate(
            input_variables=["topic"],
            template="Ask 3 questions about {topic}"
        ),
        "storytelling": PromptTemplate(
            input_variables=["topic"],
            template="Explain {topic} as a story"
        )
    }

    return templates.get(style)
```

```python
topic = "Neural Network"
print(get_prompt_by_style("teaching").format(topic=topic))
print(get_prompt_by_style("interview").format(topic=topic))
print(get_prompt_by_style("storytelling").format(topic=topic))
```

**5. Task 4: ChatPromptTemplate System`**

```python
def create_chat_prompt(role, topic):

    system_messages = {
        "teacher": "You are a helpful teacher who explains concepts clearly.",
        "interviewer": "You are an interviewer asking challenging questions.",
        "motivator": "You are a motivational speaker who inspires learning."
    }

    system_message = system_messages.get(role)

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Talk about {topic}")
    ])

    return chat_template.format_messages(topic=topic)
```

```python
messages = create_chat_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
```

**7. Task 6: Prompt Generator App**

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone
```

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    style_templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions on {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    selected_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=style_templates.get(style)
    )

    return selected_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

```python
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

**8. Task 7: Reusability Test**

```python
def reusable_template_test():

    template = PromptTemplate(
        input_variables=["topic", "audience"],
        template="Explain {topic} for {audience}"
    )

    test_data = [
        ("AI", "beginner"),
        ("Python", "student"),
        ("Data Science", "analyst"),
        ("Deep Learning", "engineer"),
        ("Statistics", "researcher")
    ]

    for topic, audience in test_data:
        print(template.format(topic=topic, audience=audience))
```
