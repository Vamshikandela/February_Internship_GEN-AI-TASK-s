**2.Pass / Fail Analyzer**


```python
marks = [45, 78, 90, 33, 60]
pass_count=0   #pass count intilize with 0
fail_count=0   #fail count intilize with 0
for mark in marks: # it will marks one by one
  if mark>=50:  # it will checks there is any marks greater or equals to 50 passing score
    pass_count+=1  # if there it will add to pass_count
  else:     # if not there
    fail_count+=1  ## it will adds to fail_count
print('-'*50,"Academic evaluation systems",'-'*50)
print("The Number of Students are passed: ",pass_count)
print("The Number of students are failed:  ",fail_count)
```

**3.Data preprocessing before analysis**



```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names=[]   # created a new list for cleaned names storing
for name in names:   # it will return all names one  by one
  names=name.strip().lower() ## It will strip() removes all empty spaces left and right side and lower() will changes all into lower cases
  cleaned_names.append(names) ## it will add al the cleaned_messages in empty  list
print("cleaned_usernames list: ",cleaned_names)  # it will display  the output
```

**4.Text filtering and validation systems**
